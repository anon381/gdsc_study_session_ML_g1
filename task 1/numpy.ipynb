```python
import numpy as np
```

```python
nums = [11, 56, 80, 39, 56]
np_nums = np.array(nums)
```

```python
mean_value = np.mean(np_nums)
max_value = np.max(np_nums)
sum_value = np.sum(np_nums)
```

```python
print(f"Numbers: {np_nums}")
print(f"Mean: {mean_value}")
print(f"Maximum Value: {max_value}")
print(f"Sum: {sum_value}")
```
