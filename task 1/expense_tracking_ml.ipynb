# Problem: Predicting Overspending in Expense Tracking

## 1️⃣ Identify an ML Problem
**Problem:** Predict whether a user will overspend next month based on their past spending habits.

## 2️⃣ Classify the ML Type
**Type:** Supervised Learning

**Reason:**
- We have labeled historical expense data (income, categories, monthly totals → overspent / not overspent).
- The model learns from examples where the correct output is already known.
- Supervised learning is used when predicting a clear labeled outcome.

## 3️⃣ Outline the ML Workflow
1. **Collect Data** – user’s monthly expenses (food, bills, transport, shopping, entertainment).
2. **Clean Data** – fix missing values, remove duplicates.
3. **Feature Engineering** – spending totals, category ratios, savings rate.
4. **Split Data** – training and test sets.
5. **Train Model** – logistic regression, random forest, etc.
6. **Evaluate** – accuracy, precision, recall.
7. **Deploy** – warn users when they are likely to overspend.

## 4️⃣ Identify One Possible Failure Point
**Failure Point:** Bad / Noisy Data

Expense data might be:
- incorrect
- missing
- placed in wrong categories
- inconsistent
